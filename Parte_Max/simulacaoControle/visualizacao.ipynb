Teste feito para um controlador PI em serie com um controlador PD

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
f = open("outputPD.txt", "r")
arrPD = f.read().split(',')
len(arrPD)
arrPD = arrPD[:len(arrPD)-2]#O ultimo carectere é um espaço em branco

arrPD = [float(x) for x in arrPD]
```

```python
fI = open("outputPI.txt", "r")
arrPI = fI.read().split(',')
len(arrPI)
arrPI = arrPI[:len(arrPI)-2]#O ultimo carectere é um espaço em branco
arrPI = [float(x) for x in arrPI]
```

```python
#Cria um degrau 
deg  = []
for i in range(0,3000):
    if i < 10:
        deg.append(0)
    else:
        deg.append(1)
```

```python
plt.plot(arrPI)
```

```python
plt.plot(arrPD)
plt.plot(deg)
```
